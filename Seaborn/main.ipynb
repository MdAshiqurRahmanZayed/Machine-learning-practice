```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.get_dataset_names()
```

```python
tips  = sns.load_dataset('tips')
iris  = sns.load_dataset('iris')
titanic  = sns.load_dataset('titanic')
planets  = sns.load_dataset('planets')
```

```python
tips.head()
```

```python
sns.scatterplot(x= 'tip',y='total_bill',data=tips,hue='day',size = 'size')
```

```python
ax = sns.scatterplot(x= 'tip',y='total_bill',data=tips,hue='day',size = 'size')
ax.set(xlabel='Tip',ylabel='Total Bill')
plt.show()
```

```python
sns.color_palette(palette='magma_r')
```

```python
sns.histplot(tips['tip'],kde=True)
```

```python
sns.displot(tips['tip'],kde=True)
```

```python
sns.barplot(x = 'sex',y = 'tip',data=tips,hue='time')
```

```python
sns.boxplot(x = 'day',y = 'total_bill',data=tips,hue='sex')
```

```python
sns.stripplot(x = 'day',y='tip',data=tips,hue='sex',dodge=True)
```

```python
sns.jointplot(x= 'tip',y='total_bill',data=tips,kind='kde',fill=True)
```
