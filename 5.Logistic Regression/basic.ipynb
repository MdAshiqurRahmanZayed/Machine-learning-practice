```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

# Load the data
df = pd.read_csv("insurance_data.csv")

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(df[['age']], df.bought_insurance, train_size=0.9)

# Create a logistic regression model
model = LogisticRegression()

# Train the model on the training data
model.fit(X_train, y_train)

# Scatter plot of the data points
plt.scatter(df.age, df.bought_insurance, marker="+", color="red", label="Data")

# Plot the decision boundary
X_boundary = np.linspace(df.age.min(), df.age.max(), 100)
y_boundary = model.predict_proba(X_boundary.reshape(-1, 1))[:, 1]
plt.plot(X_boundary, y_boundary, color="blue", label="Decision Boundary")

plt.xlabel("Age")
plt.ylabel("Bought Insurance")
plt.title("Logistic Regression Decision Boundary")
plt.legend()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

# Load the data
df = pd.read_csv("insurance_data.csv")

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(df[['age']], df.bought_insurance, train_size=0.9)

# Create a logistic regression model
model = LogisticRegression()

# Train the model on the training data
model.fit(X_train, y_train)

# Generate a range of age values for the decision boundary
X_boundary = np.linspace(df.age.min(), df.age.max(), 100)

# Predict the probabilities of class 1 for each age value
y_boundary = model.predict_proba(X_boundary.reshape(-1, 1))[:, 1]

# Plot the decision boundary
plt.plot(X_boundary, y_boundary, color="blue", label="Decision Boundary")
plt.xlabel("Age")
plt.ylabel("Probability of Buying Insurance (Class 1)")
plt.title("Logistic Regression Decision Boundary")

# Add a horizontal line at y=0.5
plt.axhline(y=0.5, color="red", linestyle="--", label="Probability = 0.5")

# Mark the intersection points
intersection_age = X_boundary[np.argwhere(np.diff(np.sign(y_boundary - 0.5))).flatten()]
intersection_prob = np.full_like(intersection_age, 0.5)
plt.scatter(intersection_age, intersection_prob, color="red", marker="o", label="Intersection")

plt.legend()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

# Load the data
df = pd.read_csv("insurance_data.csv")

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(df[['age']], df.bought_insurance, train_size=0.9)

# Create a logistic regression model
model = LogisticRegression()

# Train the model on the training data
model.fit(X_train, y_train)

# Generate a range of age values for the decision boundary
X_boundary = np.linspace(df.age.min(), df.age.max(), 100)

# Predict the probabilities of class 1 for each age value
y_boundary = model.predict_proba(X_boundary.reshape(-1, 1))[:, 1]

# Calculate the middle age value of the decision boundary (where probability = 0.5)
middle_age = X_boundary[np.argmin(np.abs(y_boundary - 0.5))]

# Plot the scatter points for age and bought insurance
plt.scatter(df.age, df.bought_insurance, marker="+", color="red", label="Data")

# Plot the decision boundary
plt.plot(X_boundary, y_boundary, color="blue", label="Decision Boundary")
plt.xlabel("Age")
plt.ylabel("Probability of Buying Insurance (Class 1)")
plt.title("Logistic Regression Decision Boundary")

# Add a vertical line at the middle of the decision boundary
plt.axvline(x=middle_age, color="red", linestyle="--", label="Middle of Decision Boundary")

plt.legend()
plt.show()

# Print the middle age value
print("Middle Age of Decision Boundary:", middle_age)
```
